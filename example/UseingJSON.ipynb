```python
import plotly.express as px
import json
import pandas as pd
df = pd.read_excel('002.xlsx',encoding="UTF-8")
```

```python
df.head()
```

```python
with open('USA.json') as jsonfile:
    GeoJSON = json.load(jsonfile)
    
fig = px.choropleth(df, geojson=GeoJSON, color="pop",
                    locations="id", featureidkey="properties.HASC_1",
                    projection="mercator"
                   )
fig.update_geos(fitbounds='locations', visible=False)
fig.show(renderer="png")
```
